Q.2)In which year the highest number of TV shows & Movies were released? Show  with bar graph.

```python
high_year=pd.DataFrame(df.groupby('type')[['release_year']].value_counts())
```

```python
high_year.rename(columns = {0:'count'},inplace = True)
```

```python
high_year.sort_values(by='count',ascending = False)
```

```python
high_year.reset_index(inplace=True)
high_year
```

```python
df_movie=high_year[high_year['type']=='Movie'].sort_values( by = 'count',ascending = False).head(10)
df_tv_show = high_year[high_year['type']=='TV Show'].sort_values(by = 'count',ascending = False).head(10)
```

```python
sns.barplot(x='release_year',y='count',data=df_movie)
plt.title('Movie Per year')
plt.show()
```

```python
sns.barplot(x='release_year',y='count',data=df_tv_show)
plt.title('tv shows Per year')
plt.show()
```
